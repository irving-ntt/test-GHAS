```python
#parametros
params = WidgetParams(
    {
        "sr_folio": str,
        "sr_fec_liq": (str, "YYYYMMDD"),
        "sr_subproceso": str,
        "sr_etapa": str,
        "sr_fec_acc": (str, "YYYYMMDD"),
        "sr_tipo_mov": str,
        "sr_etapa": str,
        "sr_instancia_proceso": str,
        "sr_usuario": str,
        "sr_id_snapshot": str,
    }
)
params.validate()
```

```python
conf = ConfManager()

#Archivos SQL
query = QueryManager()

#Conexion a base de datos
db = DBXConnectionManager()
```

```python
# Cargar query para extraer datos de ETL MOVS para insertar en CIERREN.TTAFOGRAL_MOV_VIV
 
statement = query.get_statement(
    "NB_ANCIN_DB_310_TTAFOGRAL_MOV_VIV.sql",
    sr_folio=params.sr_folio,
)

db.write_delta(f"DELTA_TRA_DB_310_TTAFOGRAL_MOV_VIV_{params.sr_folio}", db.read_data("default", statement), "overwrite")

if conf.debug:
    display(db.read_delta(f"DELTA_TRA_DB_310_TTAFOGRAL_MOV_VIV_{params.sr_folio}"))
```

```python
#INSERTA EN TTAFOGRAL_MOV_VIV

table_name = "CIERREN.TTAFOGRAL_MOV_VIV"

db.write_data(db.read_delta(f"DELTA_TRA_DB_310_TTAFOGRAL_MOV_VIV_{params.sr_folio}"), table_name, "default", "append")
```
